#### locality_name

Пропуски в колонке locality_name невозможно заполнить из текущих данных, доля таких пропусков невелика и 0,21, их можно удалить и это не повлечет за собой искажение данных.

```python
#удалим строки где есть пропуски в locality_name 
df=df.dropna(subset=['locality_name'])
```

```python
#заменим расхождения в названиях из-за которых возникают дубликаты
df['locality_name']=df['locality_name'].str.replace('поселок','посёлок')
df['locality_name']=df['locality_name'].str.replace('городской посёлок','посёлок городского типа')
df['locality_name']=df['locality_name'].str.replace('посёлок станции','посёлок при железнодорожной станции')
```

#### ceiling_height

Пропуски в колонке ceiling_height(высота потолков) заменим на медианное значение для каджой группы по этажности домов, так как застройка как правило типовая и во всех пятиэтажках высота примерно одинаковая. Можно было бы сгруппировать по году постройки дома, тк серии домов отличаются от года к году, но такой информации у нас нет

```python
df['ceiling_height'].isna().sum() #посмотрим количество пропусков в столбце
```

```python
sorted(df['ceiling_height'].unique())#посмотрим на уникальные значения в данном столбце
```

```python
df.query('ceiling_height > 14') #посмотрим, записи с высотой потолков больше 10м,
#тк  такие высокие потолки могут быть в могут быть в складских помещениях, пентхаусах,
#помещениях со вторым светом.
```

Есть аномальные высокие значения. Значения в диапазоне 24-32 вероятнее всего заполнены с ошибкой,скорее всего предполагались значения 2.4-3.2, что вполне реалистично для жилых помещений. 2,4 метра считается минимальной комфортной для проживания высотой.


```python
df['ceiling_height']=df['ceiling_height'].replace(24, 2.4) #заменим значения
df['ceiling_height']=df['ceiling_height'].replace(25.0, 2.50)
df['ceiling_height']=df['ceiling_height'].replace(26.0, 2.60)
df['ceiling_height']=df['ceiling_height'].replace(27.0, 2.70)
df['ceiling_height']=df['ceiling_height'].replace(27.5, 2.75)
df['ceiling_height']=df['ceiling_height'].replace(32.0, 3.20)
```

```python
df.query('ceiling_height in [24,25,26,27,27.5,32]') #проверим что замена удалась
```

```python
df.info()
```

Значения в диапазоне 15-23 и выше 32 - вероятно заполнены с ошибкой, таких строк всего 3. Удалим их.

```python
df=df.query('ceiling_height < 15 or 23< ceiling_height < 32 or ceiling_height.isna()') #удалим строки с аномальными значениями
```

```python
df.info()
```

```python
df.query('ceiling_height > 14') #проверить что строки удалились
```

```python
df.query('ceiling_height < 2.4') #посмотрим строки, где высота потолков аномально маленькая
```

```python
df=df.query('~(ceiling_height < 2.4)')#удалим эти строки
```

```python
for t in df['floors_total'].unique():
    df.loc[(df['floors_total'] == t) & (df['ceiling_height'].isna()), 'ceiling_height'] = \
    df.loc[(df['floors_total'] == t), 'ceiling_height'].median()
#заменили пропуски медианным значением по категориям этажности домов 
```

```python
df['ceiling_height'].isna().sum()
#проверим количество пропусков после заполнения,остался один пропуск
```

```python
df[df['ceiling_height'].isna()]  #смотрим что за строка, высота - 33 этаж
```

```python
df['floors_total'].value_counts()
#смотритм, cколько всего домов с высотой 33 этажа, такой дом всего 1 и там пропуск,
#поэтому по нему не считается медиана.
```

```python
# посчитаем медианное значение высоты потолка для домов с похожей высотой - 34,35,36,37 этажей
median_height_30s = df.query('floors_total in [34,35,36,37]')['ceiling_height'].median()
median_height_30s
```

```python
df['ceiling_height']=df['ceiling_height'].fillna(value=median_height_30s)
# заполним это значение медианным для домов с похожей высотой - 34,35,36,37 этажей
```

```python
df['ceiling_height'].isna().sum() #все пропуски заполнены
```

Заполним оставшиеся пропуски медианным значением для квартир с одинаковым количеством комнат в домах одинаковой этажности.

```python
#построим сводную таблицу по этажности домов и количеству комнат
floors_rooms_median = df.pivot_table(index = ['floors_total','rooms'], values =['living_area','kitchen_area'], aggfunc = 'median')
```

```python
floors_rooms_median
```

```python
# функция, которая принимает на вход строку датасета и если встречает пропуск в колонке(col),
#то заполняет его медианным знаечением из сводной таблицы floors_rooms_median для 
#соответствующего количества этажей и комнат

def fix_na(row, col):
    rooms = row['rooms']
    floors = row['floors_total']
    
    if pd.isna(row[col]):
        try:
            new_value = floors_rooms_median.loc[floors,rooms][col]
            return new_value
        except: 
            return row[col]
    else:
        return row[col]
```

```python
df['living_area'] = df.apply(lambda row: fix_na(row, col='living_area'), axis=1)
#применим функцию к столбцу living_area
```

```python
df['living_area'].isna().sum() # посмотрим оставшиеся пропуски
```

```python
df['living_area'] = df.apply(lambda row: fix_na(row, col='kitchen_area'), axis=1)
#применим функцию к столбцу kitchen_area
```

```python
df['kitchen_area'].isna().sum() #посмотрим оставшиеся пропуски
```

Оставшиеся значения удалим, тк нет адекватного способа заполнить их релевантными значениями, а их количество не столь велико в процентном соотношении от всего датасета.
